```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
f = open('T.txt', 'r')
data = f.read()
f.close()
nums = list(map(float,data.split()))
x = np.zeros(0)
y = np.zeros(0)
for i in range(len(nums)): 
    if i % 2 == 0: 
        y = np.append(y, nums[i])
    else :
        x = np.append(x, nums[i])
plt.figure(figsize=(10, 8))
plt.xlabel('N')
plt.ylabel('T, s')
plt.grid(True)
plt.title('T(N) for a = 0, b = 100, x = 50, p = 0.51, P = 4')
plt.plot(y, x)
plt.show()
```
